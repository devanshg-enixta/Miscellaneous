```python
import random
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
CHUNK = 1024
RECORD_SECONDS = 5
def input_reader(queue):
    while True:
        file_name = "audio_"+''.join(map(str,[random.randint(1,15) for x in range(10)]))+'.wav'
        audio = pyaudio.PyAudio()
        # start Recording
        print "Recording..."
        stream = audio.open(format=FORMAT, channels=CHANNELS,
                        rate=RATE, input=True,
                        frames_per_buffer=CHUNK)
        frames = []

        for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
            data = stream.read(CHUNK)
            frames.append(data)

        waveFile = wave.open(file_name, 'wb')
        waveFile.setnchannels(CHANNELS)
        waveFile.setsampwidth(audio.get_sample_size(FORMAT))
        waveFile.setframerate(RATE)
        waveFile.writeframes(b''.join(frames))
        waveFile.close()
        
        queue.put(file_name)
```

```python
import os
os.environ['MULTIPROCESSING_FORKING_DISABLE'] = '1' 
import multiprocessing as mp
import time
# mp.forking_enable(False)

manager = mp.Manager()
input_queue = manager.Queue()
# output_queue = manager.Queue()

try:
    process1 = mp.Process(target=input_reader, args=(input_queue,))
    process2 = mp.Process(target=speech2text, args=(input_queue,))
    # process3 = mp.Process(target=output_write, args=(output_queue))
    process1.start()
    process2.start()
    time.sleep(5.)
    process1.join()
    process2.join()
except (KeyboardInterrupt, SystemExit):
    print "Got Interrupt. Killiing the processes!"
    process1.terminate()
    process2.terminate()
```
